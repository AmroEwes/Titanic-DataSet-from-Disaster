# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

## Importing Train Data

```python
data_train = pd.read_csv('train.csv')
```

### Data Manipulation

```python
data_train.describe(include='all')
```

```python
data_train.head()
```

```python
data_train = data_train.set_index('PassengerId')
data_train
```

```python
data_train.columns
```

```python
columns = ['Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket',
       'Fare', 'Cabin', 'Embarked','Survived']
data_train = data_train[columns]
data_train.head()
```

### Dropping Unnecessary Coiumns

```python
data_train = data_train.drop('Name',axis=1)
data_train
```

```python
data_train = data_train.drop('Ticket',axis=1)
data_train
```

```python
data_train = data_train.drop('Cabin',axis=1)
data_train
```

```python
data_train.describe(include='all')
```

### Data Cleaning and Preprocessing

```python
data_train.isnull().sum()
```

```python
data_train['Age'] = data_train['Age'].fillna(value=data_train['Age'].mean())
data_train.isnull().sum()
```

```python
data_train = data_train.dropna(axis=0)
```

```python
data_train.isnull().sum()
```

```python
data_train.describe(include='all')
```

```python
data_train['Sex'].unique()
```

```python
#data_train['Sex'] = pd.get_dummies(data_train['Sex'],drop_first=True)
data_train['Sex'] = data_train['Sex'].replace(['male','female'],[1,0])
```

```python
data_train['Embarked'].unique()
```

```python
#Embarked implies where the traveler mounted from.
#There are three possible values for Embark — Southampton, Cherbourg, and Queenstown. 
#More than 70% of the people boarded from Southampton. 
#Just under 20% boarded from Cherbourg and the rest boarded from Queenstown.
#People who boarded from Cherbourg had a higher chance of survival than people who boarded from Southampton or Queenstown.
```

```python
data_train['Embarked'] = data_train['Embarked'].replace(['S','C','Q'],[0,1,2])
```

```python
data_train['Pclass'].value_counts(1)
```

```python
data_train['Family'] = data_train.Parch + data_train.SibSp
data_train
```

```python
data_train['Is_Alone'] = data_train.Family == 0
data_train
```

```python
data_train['Is_Alone'] = data_train['Is_Alone'].astype(int)
data_train
```

```python
data_train = data_train.drop('SibSp',axis=1)
data_train = data_train.drop('Parch',axis=1)

data_train
```

```python
data_train = data_train.drop('Family',axis=1)
data_train
```

```python
data_train.columns
```

```python
columns_updated = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked', 'Is_Alone','Survived']
data_train = data_train[columns_updated]
data_train
```

```python
data_train.head()
```
